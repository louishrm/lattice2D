## Hexagonal with different labelling

```python
primitive_lattice_vectors = np.array([[1,0],[1/2,np.sqrt(3)/2]])
atom_coords = np.array([[0,0], [1/3,1/3]])
supercell_coefficients = np.array([4,1,0,3])
lattice = Lattice2D(primitive_lattice_vectors, atom_coords, supercell_coefficients)
```

```python
lattice.plot_supercell()
```

```python
sites = np.zeros((lattice.M,2))
site_index = 0
for i in range(lattice.nuc):
    sites[site_index] = lattice.periodic_site(np.array((i,0)))
    sites[site_index+1] = lattice.periodic_site(np.array((i+1/3,1/3)))
    site_index +=2
```

```python
lattice.plot_supercell(sites=sites)
```

```python
np.where(lattice.distance_matrix(sites=sites)[0]==2)
```
